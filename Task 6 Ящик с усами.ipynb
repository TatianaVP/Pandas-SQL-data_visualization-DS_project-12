#### 1. Используйте данные из этого файла, считайте их в датафрейм и внесите любые изменения, которые, по вашему мнению, помогут решить задачу.

```python
df = pd.read_csv('ab-test.csv')
```

```python
df
```

```python
df_before = df.loc[df['time'] == 'before'].reset_index(drop=True)
```

```python
df_after = df.loc[df['time'] == 'after'].reset_index(drop=True)
```

```python
df_before.rename(columns={'diff': 'diff_before'}, inplace=True)
```

```python
df_after.rename(columns={'diff': 'diff_after'}, inplace=True)
```

```python
df_for_plot = df_before.join(df_after['diff_after']).drop(['uid', 'time'], axis=1)
```

```python
df_for_plot
```

#### 2. Размер figsize остается прежним, размер шрифта можно выбрать по своему вкусу.
#### 3. Цветовая палитра должна быть такой же, как в примере.
#### 4. Размер шрифта заголовка — 15.
#### 5. Ширина линий ящиков — 3, ширина линий медианы — 2.

```python
df_for_plot.boxplot(by='group', figsize=(15, 8), boxprops={'linewidth': 3}, medianprops={'linewidth':2},showmeans=True, whiskerprops={'color':'red'}, capprops={'color': 'red'})
plt.suptitle('Lag before labs deadlines. After and before the Newsfeed ', size=15)
```
